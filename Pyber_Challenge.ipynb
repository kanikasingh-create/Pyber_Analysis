```python
# Add Depedencies
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pandas as pd
import numpy as np
```

```python
# Part 1
```

```python
# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# Read the city data file and store it in a pandas DataFrame.
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
# Read the ride data file and store it in a pandas DataFrame.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.head(10)
```

```python
# Combine the data into a single dataset.
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

pyber_data_df.head()
```

```python
# Set Index to 'Type'.
pyber_type = pyber_data_df.set_index(["type"])
pyber_type
```

```python
# Total Rides
rides_summary = pyber_type.groupby(["type"]).count()["ride_id"]
rides_summary.head()
```

```python
# Total Drivers
drivers_summary = pyber_type.groupby(["type"]).sum()["driver_count"]
drivers_summary.head()
```

```python
# Total Fares
fare_summary = pyber_type.groupby(["type"]).sum()["fare"]
fare_summary.head()
```

```python
# Create the summary DataFrame with the appropriate columns and apply formatting where appropriate.
pyber_type_summary_df = pd.DataFrame({"Total Rides": rides_summary, 
                                     "Total Drivers": drivers_summary, 
                                     "Total Fares": fare_summary.map("${:,.2f}".format),
                                     "Average Fare per Ride": (fare_summary/rides_summary).map("${:,.2f}".format),
                                    "Average Fare per Driver": (fare_summary/drivers_summary).map("${:,.2f}".format)})

#Delete Index.
pyber_type_summary_df.index.name = None

#Show DataFrame.
pyber_type_summary_df
```

```python
# Part 2
```

```python
# Rename columns in merged DataFrame.
pyber_renamed=pyber_data_df.rename(columns={'city': 'City', 'date':'Date','fare':'Fare', 'ride_id': 'Ride Id','driver_count': 'No. Drivers', 'type':'City Type'})

pyber_renamed.head()
```

```python
# Set index to the date column.
pyber_dates = pyber_renamed.set_index(["Date"])
pyber_dates.head()
```

```python
#Create a new DataFrame for fares and include only the Date, City Type, and Fare columns using the copy() method on the merged DataFrame.
pyber_copy=pyber_renamed.copy()
date_formatted=pd.to_datetime(pyber_copy["Date"]) 
datetime_index = pd.DatetimeIndex(date_formatted.values)
pyber_copy1=pyber_copy.set_index(datetime_index)
pyber_copy1
```

```python
pyber_copy1.info()
```

```python
# Drop the extra Date column.
pyber_formatted=pyber_copy1.drop(columns=["Date"])

#Set the index to the datetime data type.
pyber_formatted.index.name = "Date"

# Check the DataFrame using the info method to make sure the index is a datetime data type.
pyber_formatted.info()
```

```python
# Calculate the sum of fares by the type of city and date using groupby.
fare_date_summary = pyber_formatted.groupby(["Date", "City Type"]).sum()["Fare"]
fare_date_summary.head()
```

```python
# Create a new DataFrame with above.
pyber_faredate_summary_df = pd.DataFrame({"Total Fares": fare_date_summary})
pyber_faredate_summary_df
```

```python
# Create a pivot table DataFrame, with the Fare for each Date in each row. 
type_pivot=pd.pivot_table(pyber_faredate_summary_df,values=["Total Fares"], index=["Date"], columns=['City Type'],aggfunc=np.sum)
type_pivot
```

```python
# Create a new DataFrame from the pivot table DataFrame using loc.
pivot_dates = type_pivot.loc['2019-01-01':'2019-04-28']
pivot_dates
```

```python
# Create a new DataFrame by setting the DataFrame created above in weekly bins, and calculate the sum() of the fares for each week.
resample_dates=pivot_dates.resample('7D').sum()
resample_dates
```

```python
# Plot DataFrame.
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots()


fig.set_size_inches(18.5, 10.5)
ax.plot(resample_dates)
fig.autofmt_xdate()
ax.fmt_xdata = mdates.DateFormatter('%Y-%m-%d')
plt.xlabel('')
plt.ylabel('Fare $USD')
plt.legend(loc="center")
plt.title("Total Fare by City Type")
plt.savefig("analysis/Challenge.png")
plt.show()
```
